```python
ftse100 = yf.Ticker("^FTSE")
ftse100_history = ftse100.history(period="180d")
ftse = ftse100_history
ftse.head(10)
```

```python
from nalyst.ThresholdClassifier import ThresholdClassifier, ThresholdPlot
```

```python
classifier = ThresholdClassifier(0.5)
```

```python
classification = classifier.classify(ftse['Close'].tolist())
plot = ThresholdPlot(ftse['Close'], classification)
plot.plot()
```

```python
anglo_american = yf.Ticker("AAL.L")
aal_history = anglo_american.history(period="180d")
aal = aal_history
aal.head(10)
```

```python
classification = classifier.classify(aal['Close'].tolist())
plot = ThresholdPlot(aal['Close'], classification)
plot.plot()
```

```python
from nalyst.TestTrainSplit import TrainTestSplit
```

```python
X = ftse.Close.values.reshape(-1,1)
y = aal.Close.values.reshape(-1,1)
```

```python
X.shape
```

```python
y.shape
```

```python
X = X[1:]
X.shape
```

```python
X[:5]
```

```python
y[:5]
```

```python
splitter = TrainTestSplit(X, y, test_size=0.30)
```

```python
X_train, X_test, y_train, y_test = splitter.split()
```

```python
print(len(X_train))
print(len(y_train))
```

```python
print(len(X_test))
print(len(y_test))
```

```python
from nalyst.LinearRegression import LinearRegression
```

```python
# create the LinearRegression object
regression_model = LinearRegression(X, y)
```

```python
# make predictions for x values 
predictions = regression_model.predict(X_test)
print("Predictions: ", predictions[:5])
```

```python
intercept = regression_model.bo
print("Intercept",intercept[0])
```

```python
slope = regression_model.b1
print("Slope:",slope[0])
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(15,6))
plt.scatter(X,y, color="red")
plt.plot(X_test, predictions, color="white")
plt.grid(linestyle='--', color='gray', alpha=0.8)
plt.title("Regression Model (FTSE100 V/S Anglo American)")
plt.ylabel("Anglo American plc")
plt.xlabel("FTESE100")
plt.show()
plt.close()
```

```python
from nalyst.RegressionPlot import RegressionPlot
```

```python
plot = RegressionPlot.residual_plot(regression_model)
```

```python
model = LinearRegression(X, y)

# Calculate errors
mse = model.mean_squared_error()
rmse = model.root_mean_squared_error()
r_squared_value = model.r_squared()
```

```python
# Print results
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r_squared_value}")
```

```python
from nalyst.BetaFive import calculate_beta_five
from nalyst.BetaMax import calculate_beta_max
```

```python
df = pd.read_excel('sample_data/ftse.xlsx')
df.tail()
```

```python
df.columns = df.columns.str.lower()
df['ticker'] = df['ticker'].apply(lambda x: x + '.L')
```

```python
df[:5] #L stand for London Stock Exchange 
```

```python
companies = df.ticker.unique()
companies[:20]
```

```python
ftse_100 = []
ticker = df.ticker.unique()
for i in ticker:
    ftse_100.append(i)
```

```python
ftse_100[:5]
```

```python
type(ftse_100) #list type
```

```python
tickers = ftse_100
benchmark = "^FTSE"

i = 1
for ticker in tickers:
    beta = calculate_beta_max(ticker, benchmark)
    print(i, "The beta of " + ticker + " is: " + str(beta))
    i += 1
```
